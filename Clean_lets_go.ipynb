```python
t1 = pd.read_csv("Lego_processed.csv",encoding = 'latin1')
```

```python
t1[t1.user_type.str.len()<=3]
```

```python
t1["system_type"] = t1["system_type"].str.lstrip().str.rstrip()
t1.loc[t1["system_type"].str.len()==0,"system_type"] = 'none'
t1.loc[t1["system_type"]=="","system_type"] = 'none'
t1["system_type"] = t1["system_type"].fillna('none') 
```

```python
t1[t1.system_type.isna()==True]
```

```python
uid_prev = t1.loc[0,"user_id"]
bot_prev = ""

uid_list = []
bot_list = []
cust_list = []
fg=0
for i,row in t1.iterrows():
    uid = row["user_id"]
    bot_txt = row["system"]
    cust_txt = row["system_type"]
    
    if(uid_prev!=uid):
        uid_prev= t1.loc[i,"user_id"]
        bot_prev=""
        fg=0

    if(uid_prev==uid) & (cust_txt=="none"):
        bot_txt = bot_prev + bot_txt
            
        fg=1
    else:
        if(cust_prev=="none") & (fg==1)& (uid_prev==uid) :
            bot_txt = str(bot_prev) + bot_txt
            fg=0
    #if(i>=23) & (i<=27):
        #print (inp_txt,fg)
            
    if((i==0)|(uid_prev==uid)):
        uid_prev = uid
        bot_prev = bot_txt
        cust_prev=cust_txt
        

        
    if(fg==0):
        uid_list.append(uid)
        bot_list.append(bot_txt)
        cust_list.append(cust_txt)
            
    
            
        
```

```python
df = pd.DataFrame(uid_list)
df.columns = ["user_id"]
df["bot_text"] = bot_list
df["cust_text"] =cust_list
```
